```python
df["date"] = pd.to_datetime(df["date"])
df = df[(df["airline"] == "IndiGo") & 
    (df["date"].dt.year ==2025) & 
    (df["date"].dt.month == 12) & 
    (df["cancelled_flights"] > 0)]
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
original_df = df.copy()
```

```python
conditions = [df["cancelled_flights"] >= 3,
        df["cancelled_flights"] >= 2,
        df["cancelled_flights"] == 1,
    ]

choices = ["High", "Moderate", "Low"]

df["crisis_severity"] = np.select(
    conditions,
    choices,
    default="Low"
)

df[["cancelled_flights", "crisis_severity"]].head()
```

```python
df["fare"] = np.select(
    [
        df["avg_fare_inr"] >= 7000,
        df["avg_fare_inr"] >= 3000,
    ],
    [
        "High",
        "Medium"
    ],
    default="Low"
)
```
